```python
# outlier detection can be for columns - 
outlier_columns = ['Waste Generated (Tons/Day)', 'Population Density (People/km²)',
       'Cost of Waste Management (₹/Ton)', 
       'Landfill Capacity (Tons)', 'Year']
for i in outlier_columns:
    sns.boxplot(df[i])
    plt.xlabel(i)
    plt.ylabel('count')
    plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder
col = ['City/District', 'Waste Type', 'Disposal Method', 'Landfill Name']
df = pd.get_dummies(df, columns=col, drop_first=True, dtype='int')
df
```

```python
df[['Latitude', 'Longitude']] = df['Landfill Location (Lat, Long)'].str.split(',', expand=True)

df['Latitude'] = df['Latitude'].astype(float)
df['Longitude'] = df['Longitude'].astype(float)

df.drop('Landfill Location (Lat, Long)', axis=1, inplace=True)
```
